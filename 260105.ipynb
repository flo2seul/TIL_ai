```python
A = set(["park", "kim", "lee", "na"])
B = set(["park", "choi", "cho", "gang", "na"])

#common = []

# for attendee in A:
#     if attendee in B and attendee not in common:
#         common.append(attendee)

common = A & B # 교집합 A.intersection(B)
common = A | B # 합집합 A.union(B)
common = A - B # 차집합 A.difference(B)

print("2개의 행사에 모두 참여한 사람은?")
print(common)

print(A.intersection(B))
```
