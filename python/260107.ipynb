```python
class Treasures:

    __slots__ = ["account", "signature"]
    __balance = 0
    __savings = [0, 0, 0]

    def __init__(self, **kwargs):

        self.account = kwargs["account"]
        self.signature = kwargs["signature"]
        Treasures.__balance += kwargs["reward"]

    def accumulate(self, reward):

        Treasures.__savings[0] = Treasures.__balance
        Treasures.__savings[1] = reward
        Treasures.__savings[2] = Treasures.__balance + reward

        Treasures.__balance = Treasures.__savings[2]

        return Treasures.__savings

hgd = Treasures(**{"account": "David", "signature": "D", "reward": 50000})

reward = 270000
savings = hgd.accumulate(reward)

print(f"before : {savings[0]:>10,}") #intager : >,


print(f"reward : {savings[1]:>10,}")
print(f"after : {savings[2]:>10,}")

print(f"{hgd.account}님, 상급 {savings[1]:,} 달란트가 누적되어 " + f"상급은 총 {savings[2]:,} 달란트가 되었습니다.")
```
