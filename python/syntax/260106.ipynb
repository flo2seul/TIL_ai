```python
def calc(x,y,z):
    return (x+y)*z

result = calc(z=30, x=10, y=20)
print(result)
```

```python
def print_sum(*nums, sumfmt="Sum is %d"): #nums는 튜플 변수로 정의됨

    sum = 0
    for n in nums:
        sum += n
    print(sumfmt % sum)

print_sum()
print_sum(1,2,3,4,5)
print_sum(1,2,3,4,5, sumfmt="전체 합은 %d 입니다.")
```

```python
#map
def square(n):
    return n*n
lst = [1,2,3,4,5]

lst2 = map(square, lst)
sq_lst = list(lst2)
print(sq_lst)
```

```python
#filter
lst = [1,2,3,4,5]

def less3(n):
    if n < 3: return True
    else: return False

lst2 = filter(less3, lst)
sq_lst = list(lst2)
print(sq_lst)
```

```python
#lambda
lst = [1,2,3,4,5]

lst2 = map(lambda x: x*x, lst)
sq_lst = list(lst2)
print(sq_lst)
```
