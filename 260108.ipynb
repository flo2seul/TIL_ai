```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

s = pd.Series(np.random.randn(20), index=np.arange(1,21))
s.plot()

s.plot(kind="bar")

s = pd.Series(np.random.randn(100))
s.hist(bins=50)

df = pd.DataFrame((np.random.rand(3,4)*100).round(), index=['2023','2024','2025'], columns=['Perl','C','Java','PHP'])

df.plot(kind='bar')
plt.show()
```

```python
import seaborn as sns

#sns.set_style('whitegrid')
df = pd.DataFrame(np.random.normal(80,5,300).reshape(100,3),columns=['Python','C','Java'])
sns.distplot(df['Java'],hist=True,kde=True,rug=True,bins=30)

plt.show()
```

```python
import seaborn as sns
tips = sns.load_dataset('tips')
df = pd.DataFrame(tips.groupby('day').tip.sum())
plt.bar(df.index, df['tip'], color='red', alpha=0.3, width=0.5)
# plt.barh(df.index, df['tip'], color='red', alpha=0.3, height=0.5)
plt.title('Tip stats by day', fontsize=17)
plt.xlabel('days', fontsize=13)
plt.ylabel('value', fontsize=13)
plt.xticks(df.index, fontsize=13, rotation=90)
plt.show()
```

```python
x = abs(tips['total_bill'])
y = abs(tips['tip'])
plt.scatter(x, y,
cmap='Spectral',
c=tips['tip'],
edgecolor='black',
linewidth=0.2,
alpha=0.9,
s=tips['total_bill']*5)
plt.colorbar(label='Tip')
```

```python
tips = sns.load_dataset('tips')
sns.relplot(data=tips, x='total_bill', y='tip',
col='time', hue='smoker', size='size')
```

```python
sns.boxplot(x='day', y='total_bill', hue='smoker',
palette='Blues_d', orient='v', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='.25')
```

```python
sns.jointplot(x='total_bill', y='tip', kind='kde',
space=0, color='r', data=tips)
```
