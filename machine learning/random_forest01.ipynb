```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from IPython.display import display
from sklearn.model_selection import train_test_split


loan_data = pd.read_csv('../datasets/decision/train_loan_80.csv')
display(loan_data.head())
loan_encoded = pd.get_dummies(loan_data, drop_first=True)
display(loan_encoded.head())
x = loan_encoded.drop(columns=['Loan_Status_Y'])
y = loan_encoded['Loan_Status_Y']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

# tree 200개
model = RandomForestClassifier(n_estimators=200, max_depth=5, random_state=42)
model.fit(x_train, y_train)
print(f'학습 데이터:{model.score(x_train, y_train)}')
print(f'테스트 데이터:{model.score(x_test, y_test)}')
```

```python
#중요도 값 추출
importances_loan = model.feature_importances_
#상위 5개
top_5_idx = np.argsort(importances_loan)[::-1][:5]
top_5_df = pd.DataFrame({'평가 기준feature':x.columns[top_5_idx]
                        ,'결정 영향도 : ':importances_loan[top_5_idx]})
display(top_5_df)
```
