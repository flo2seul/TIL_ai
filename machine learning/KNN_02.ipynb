```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from IPython.display import display
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
import numpy as np

np.random.seed(42)
wine = datasets.load_wine()
x = wine.data
y = wine.target
target_nm = wine.target_names
feature_nm = wine.feature_names
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
#스케일링 전 측정
knn_unscaled = KNeighborsClassifier(n_neighbors=5)
knn_unscaled.fit(x_train, y_train)
print(f'[스케일링 전] knn 정확도(k=5):{knn_unscaled.score(x_test, y_test)}')
df = pd.DataFrame(x, columns=feature_nm)
display(df.describe())
#스케일링
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
knn_scaled = KNeighborsClassifier(n_neighbors=5)
knn_scaled.fit(x_train_scaled, y_train)
print(f'[스케일링 후] knn 정확도(k=5):{knn_scaled.score(x_test_scaled, y_test)}')
```
