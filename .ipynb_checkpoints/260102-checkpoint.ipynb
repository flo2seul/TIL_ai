```python
num = int(input("몇명을 넣으시겠습니까?"))
info_list = []

for i in range(num):

    
    collect_list = []
    collect_list.append(input("{}번째 사람의 이름을 입력하세요:".format(i+1)))
    collect_list.append(input("{}번째 사람의 주소를 입력하세요:".format(i+1)))
    collect_list.append(input("{}번째 사람의 성별을 입력하세요:".format(i+1)))
    collect_list.append(input("{}번째 사람의 나이를 입력하세요:".format(i+1)))

    info_list.append(collect_list)


print(info_list)

print("이름 주소 성별 나이")
```
